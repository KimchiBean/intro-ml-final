This project will analyze the comments on the top 100 most popular recipes on www.tasteofhome.com and make predictions on how well a recipe ranks based on the comments. In exploring this, we seek to determine how well-correlated the activity on the recipe post and the success of the recipe are. There are likely recommender algorithms on this site that promote recipes with high engagement. It's also possible that the top 100 ranking is generated by an algorithm rather than being hand-picked. In theory, this model could be used to provide insight into how engagement could be manipulated into rigging the rankings.

We will begin with applying a basic, multivariable linear model. Then, we will examine other models, such as decision trees, ensembles, and support vector machines, to see if any have particular strengths in accuracy or processing speeds. As the goal is to predict a ranking, which is a continuous variable, we will be applying regression techniques.

```python
# importing all the required libraries

from math import exp
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor  
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import PredictionErrorDisplay
```

The data is sourced from a bigger study on sentiment analysis on recipe reviews:

Amir Ali, Stanisław Matuszewski, Jacek Czupyt, Usman Ahmad. (2023). Textual Taste Buds: A Profound Exploration of Emotion Identification in Food Recipes through BERT and AttBiRNN Models. International Journal of Novel Research and Development. Volume 8. ISSN: 2456-4184.

The dataset was donated to UC Irvine's machine learning repository. It has been made available on their website:

Recipe reviews and user feedback. UCI Machine Learning Repository. (n.d.). https://archive.ics.uci.edu/dataset/911/recipe+reviews+and+user+feedback+dataset 

```python
# Load the data
df = pd.read_csv('recipe_reviews.csv')

df.info()
```

All of the data is provided in a single table. The original data source contains 18182 rows, each representing 1 comment on a top 100 recipe. There are a variable amount of comments per recipe, but exactly 100 recipes represented in the data. Each of the comments has the following 15 features:

num_records: numeric; the current count of comments for a given recipe (for recipe 1, first comment is assigned a 1, second a 2, etc...)
recipe_number: numeric; the ranking of the recipe that the comment was made on
recipe_code: categorical; the ID of the recipe
recipe_name: categorical; the name of the recipe
comment_id: categorical; the ID of the comment
user_id: categorical; the ID of the user who commented
user_name: categorical; the username of the user who commented
user_reputation: numeric; the rating of the user who commented
created_at: numeric; the time when the comment was created
reply_count: numeric; the number of replies to the comment
thumbs_up: numeric; the number of upvotes on the comment
thumbs_down: numeric; the number of downvotes on the comment
stars: numeric; the rating given to the recipe by the commenter (may have not left a rating)
best_score: numberic; score given to the comment by an internal algorithm
text: categorical; the text content of the comment

There are many steps that need to be taken to clean this data. On the first iteration (cell below), the following are performed:

1. Two of the rows were missing 'text', so they have been dropped.
2. The first feature was missing a column name, so one has been added (in accordance with the features names listed in the prior section).
3. The type of 'recipe_code' has been changed from numeric to categorical

```python
# First column was missing a name in the CSV
df.rename(columns={'Unnamed: 0':'num_records'}, inplace=True )

# Remove 2 rows with missing 'text' values
df = df.dropna()

# Change incorrect inferred data types
df = df.astype({"recipe_code": object})

df.info()
```

On the second iteration, unneeded columns are removed.

These 5 columns are removed because they are effectively just IDs and distinguishers.
1. num_records
2. recipe_code
3. recipe_name
4. user_id
5. comment_id

This column is removed because, unfortunately, there is no data for when a comment was posted relative to the creation of the recipe or the creation of the rankings. So any time-based correlations cannot be made from this data alone.
1. created_at

```python
# Drop unnecessary columns
df = df.drop(columns=['recipe_code', 'recipe_name', 'user_id', 'comment_id', 'num_records', 'created_at'])
```

On the third iteration, rows with invalid values will be removed. These include the following:

1. 'text' containing the value '??', as this is effectively a NULL value
2. 'stars' being 0, as the author of the dataset claims a 0 meaning that no rating was given. We are not applying any models to guess a rating from the contents of the comment, so it will be impossible to replace the 0's with inferred values. Keeping the 0's will also skew the model since they will get interpreted as bad scores. Therefore, rows with a '0' star rating should be dropped.

Below is an example of a comment that meets both of these drop conditions.

```python
# Grab example of comments made by 'lnzpooper'
df_test = df.loc[df['user_name'] == 'lnzpooper'] 
display(df_test)
```

```python
# Drop if 'text' = ?? or 'stars' = 0
df = df.drop(df[df['text'] == '??'].index)
df = df.drop(df[df['stars'] == 0].index)

df_test = df.loc[df['user_name'] == 'lnzpooper'] 
display(df_test)
```

After dropping, running the same query results in 0 hits.

On the fourth iteration we'll focus on removing duplicate entries. In fact, the previous example also had this issue. Since that example has been dropped, we will use another to visualize the drop. Note this user has 3 reviews in the dataset, 1 of which is a duplicate. So we expect only 2 reviews to remain after the drop.

```python
# Some comments were duplicated (see lnzpooper has 4 comments for the same recipe, 2 of which have unknown text and 1 of which has duplicate text)
df_test = df.loc[df['user_name'] == 'BEVVVV'] 
display(df_test)
```

```python
# Remove duplicates (same user has multiple comments with the same text on the same recipe)
df = df.drop_duplicates(subset=['recipe_number', 'user_name', 'text'])
df_test = df.loc[df['user_name'] == 'BEVVVV'] 
display(df_test)
```

As expected, there are only 2 remaining entries for this user.

The user_name column is also an irrelevant ID. It was only kept until this point to be used for removing duplicates. It can now be dropped.

```python
# Can also drop the user_name field now that we've gotten rid of duplicates
df = df.drop(columns=['user_name'])
```

It would be interesting to train a model on the text contents of the comments to determine sentiment or maybe identify keywords common across successful recipes. This will not be covered in the scope of this project, but perhaps could be explored in the future. For now, we will instead create a new feature for the length of each comment to see if shorter or longer comments are more indicative of higher rankings.

```python
# Replace 'text' with 'text_len'
df.loc[:, 'text_len'] = df.loc[:, 'text'].str.len()
df = df.drop(columns=['text'])

df.head()
```

At this point, all of the cleaning and most of the transforming has been completed. Let's get a summary of the data now:

```python
# Get post-cleaning report
df.info()
```

The number of features has been reduced from 15 down to 8. In addition, the number of rows (comments) has been reduced from 18182 to 16230.

So far, it seems there is still a large amount of data to train the model against. However, in the current state, the problem statement cannot be handled. Since the goal is to predict rankings of recipes, having individual comments as our data points is not sufficient for training. The data must be aggregated relative to each recipe. We can see that this is true by observing the correlation matrix of the current dataset.

```python
# Check the correlation matrix
plt.figure(figsize=(15, 11))

ax = sns.heatmap(df.corr(), annot=True, fmt=".2f", linewidth=1, cmap="viridis")
ax.xaxis.tick_top()
ax.tick_params(axis='x', rotation=90)
```

As can be observed, the ranking of a recipe (recipe_number) is not correlated with any of the other features. This makes sense, as any singular comment is not indicative of how good a recipe is (imagine a recipe with 99 positive comments/ratings and 1 negative one, but only being provided the negative one to make a prediction). We need to aggregate the data.

```python
# We want to predict the recipe's ranking based on the comments on the recipes. Stats need to be aggregated per recipe
df_grouped = df.groupby('recipe_number').agg(num_comments=('recipe_number', 'count'), 
                                             user_reputation=('user_reputation', 'mean'), 
                                             reply_count=('reply_count', 'mean'), 
                                             thumbs_up=('thumbs_up', 'mean'), 
                                             thumbs_down=('thumbs_down', 'mean'), 
                                             stars=('stars', 'mean'), 
                                             best_score=('best_score', 'mean'), 
                                             text_len=('text_len', 'mean')).reset_index()
display(df_grouped)
```

After performing the aggregation, the existing features now all represent the mean of the original features (within the scope of a single recipe). For example, a recipe with a 1-star and a 5-star comment will now be represented by a single datapoint with a 3-star rating (the mean of 1 and 5). In addition, an extra feature has been added.

num_comments: numeric; number of comments on the recipe

Let's check out the correlation matrix now.

```python
# Check the correlation matrix
plt.figure(figsize=(15, 11))

ax = sns.heatmap(df_grouped.corr(), annot=True, fmt=".2f", linewidth=1, cmap="viridis")
ax.xaxis.tick_top()
ax.tick_params(axis='x', rotation=90)
```

```python
# Let's observe relationships between factors
sns.pairplot(df_grouped.iloc[:, :10], diag_kind="kde")
```

We can see now that 'recipe_number' has some correlations. Most notably, there is a high correlation with 'num_comments'. We can infer that 'num_comments' will be our best predictor in the linear regression. In addition, there is a decent level of correlation with 'stars' and 'best_score'.

These generally align with what one might logically assume. There are negative correlations with 'num_comments' and 'stars'. Meaning, as the number of comments or the rating go up, the ranking goes down. A lower numeric ranking is a better one. We expect better ratings and higher rates of engagement to be indicative of better recipes.

What may be unexpected though is the other correlation. As the average score of a comment goes up, so does the ranking. In other words, a recipe performs worse when comments, rather than recipes, are well-rated. This may suggest that more popular recipes have less substantive/meaningful comments. Perhaps comments from bots or less-passionate users are inflating a recipe's ranking?

To support this theory, we can look at some additional correlations. 'best_score' is very strongly correlated with 'thumbs_up' and 'thumbs_down'. It also has correlation with 'user_reputation' and 'reply_count'. In all cases, these correlations are positive. This means that comments with high engagement, regardless of whether this engagement is positive or negative, raises the 'best_score'.

In summary, we can infer that recipes with more comments and higher ratings tend to get higher rankings. But the average comment on these higher-ranked recipes has less community engagement. So it can be assumed that the active community is not as involved with these recipes.

Now that we've cleaned and explored the data, it's time to start training models. We'll start basic with a linear regression using the best features identified from the correlation matrix.

```python
# Create a train and test data set
X_train, X_test = train_test_split(df_grouped, test_size=0.2, random_state=42)
```

```python
# Let's train a simple linear regression
simple_linear_reg = smf.ols(f'recipe_number ~ num_comments', X_train).fit()

print(simple_linear_reg.summary())
```

Using only the single best feature identified, we achieved an adjusted R2 of 0.468. This is not terrible, but we could do better. Let's try adding the 2 next-best features, 'stars' and 'best_score'. Note that we can safely add both of these as they have low colinearity.

```python
# Adding additional features
simple_linear_reg = smf.ols(f'recipe_number ~ num_comments + stars + best_score', X_train).fit()

print(simple_linear_reg.summary())
```

The adjusted R2 increased as expected. However, despite the low colinearity value from the correlation matrix, there was colinearity introduced into the model. 'stars' has a relatively high P-value, suggesting it is colinear with the other features. We should remove it.

```python
# Adjusting for colinearity
simple_linear_reg = smf.ols(f'recipe_number ~ num_comments + best_score', X_train).fit()

print(simple_linear_reg.summary())
```

The adjusted R2 dropped slightly, but at 0.522 it is still an improvement over the original model and without colinearity issues. Let's check the leverage and residuals to see if any outliers can be pulled out of the dataset.

```python
sm.graphics.plot_leverage_resid2(simple_linear_reg)
```

The rank 2 recipe stands out as having a combination of high residual and leverage. Let's try removing it from the test data.

```python
leverage_residual_adjusted_train = X_train.drop([2])
simple_linear_reg = smf.ols(f'recipe_number ~ num_comments + best_score', leverage_residual_adjusted_train).fit()

print(simple_linear_reg.summary())
```

After removing the outlier, the adjusted R2 has increased to 0.584. So, through some basic optimization of our linear regression model, we've improved the adjusted R2 by a total of 0.116. However, we may be able to get better results from more advanced models.

Let's start by regenerating our train and test sets with the same seed, but this time also saving our Y values. We'll need them for these next models.

```python
# Create a new train and test set using the same seed and extracting the recipe_number as the Y value
y = df_grouped['recipe_number'].to_numpy()
x = df_grouped.drop(columns=['recipe_number']).to_numpy()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Below, we've defined a suite of many different models and hyperparameters. These will all be run through grid search cross validation to find the best hyperparameters for each model. 

```python
models = [
    {
        'name': 'Decision Tree',
        'model': DecisionTreeRegressor(random_state = 0),
        'params': {
            'max_depth': [None, 3, 5, 7],
            'min_samples_split': [2, 5, 10]
        }
    },
    {
        'name': 'Gradient Boost',
        'model': GradientBoostingRegressor(random_state = 0),
        'params': {
            'n_estimators': [50, 100, 150],
            'max_depth': [None, 3, 5],
            'min_samples_split': [2, 5, 10]
        }
    },
    {
        'name': 'Ada Boost',
        'model': AdaBoostRegressor(estimator=DecisionTreeRegressor(), random_state = 0),
        'params': {
            'n_estimators': [100, 200, 300],
            'estimator__max_depth': [None, 3, 5, 7],
            'estimator__min_samples_split': [2, 5, 10]
        }
    },
    {
        'name': 'Bagging',
        'model': BaggingRegressor(estimator=DecisionTreeRegressor(), random_state = 0),
        'params': {
            'n_estimators': [10, 20, 30],
            'estimator__max_depth': [None, 3, 5, 7],
            'estimator__min_samples_split': [2, 5, 10]
        }
    },
    {
        'name': 'Random Forest',
        'model': RandomForestRegressor(random_state = 0),
        'params': {
            'n_estimators': [100, 200, 300],
            'max_depth': [None, 3, 5, 7],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    {
        'name': 'Support Vector',
        'model': SVR(),
        'params': {
            'kernel': ['linear', 'poly', 'rbf'],
            'C': [1, 5, 10],
            'degree': [2, 3, 5]
        }
    }
]
```

Now lets run these models. We'll also keep track of the average runtime for each model type to compare later on.

```python
best_models = {}

# Iterate through each model type and run it using each combination of hyperparameters
for model in models:
    grid_search = GridSearchCV(estimator=model['model'], param_grid=model['params'])
    grid_search.fit(x_train, y_train)

    # Record the best results
    best_models[model['name']] = {
        'params': grid_search.best_params_,
        'y_pred': grid_search.predict(x_test)
    }

    mean_fit_time = grid_search.cv_results_['mean_fit_time']
    mean_score_time = grid_search.cv_results_['mean_score_time']

    # Calculate and print the average runtime of each model
    runtime = np.mean(mean_fit_time + mean_score_time)
    print(f"{model['name']} completed (avg {runtime} seconds per model)")
    
```

Note that the model with the best score is not necessarily the best due to potential bias and variance issues. Howevever, for the sake of simplicity in comparing different estimators, we will look at only the models that generated the best score. The best-performing hyperparameters are automatically selected by GridSearchCV.

We can display the adjusted R2, a plot of the residuals, and the hyperparameters that led to the highest score.

```python
# Iterate through each best recorded model
for key, value in best_models.items():
    # Print the best hyperparameters found and the adjusted R2 of the model run with them
    print(f"\nBest model for {key}:")
    print(value['params'])
    print(f"Adjusted R2: {r2_score(y_test, value['y_pred'])}\n")

    # Display a graph of the residuals to detect any patterns
    residuals = PredictionErrorDisplay(y_true=y_test, y_pred=value['y_pred'])
    residuals.plot()
    plt.show()
```

None of the residual plots show any discernable patterns. So there are no issues with skewed fits.

Bagging produced the best adjusted R2 at 0.805. This is a significant improvement from the linear model that produced 0.584. In fact, all of the ensemble and tree models performed fairly well with scores above 0.75. 

There are notable differences in time though. While Random Forest and Ada Boost produced scores pretty similar to Bagging (0.787 and 0.793), they both took 7x longer to run. Meanwhile, Decision Tree also got a very high score (0.791) but requires only 10% of the time needed to train the Bagging model. This dataset is fairly small after performing aggregation, so we are able to select whichever model scored the highest regardless of the time taken. But if we expanded this dataset to the last few decades and across many different recipe websites, it is easy to imagine how using Decision Trees can be favorable despite scoring slightly worse.

The use of GridSearchCV should have taken care of overfitting by performing cross validation. But let's double check by also predicting the model using the training set to see how it performs.

```python
# Check for overfitting
best_regressor = BaggingRegressor(estimator=DecisionTreeRegressor(max_depth=3, min_samples_split=2), random_state = 0, n_estimators=30)
best_regressor.fit(x_train, y_train)

y_pred_train = best_regressor.predict(x_train)
y_pred_test = best_regressor.predict(x_test)

print(f"Adjusted R2 for Training Data: {r2_score(y_train, y_pred_train)}")
print(f"Adjusted R2 for Test Data: {r2_score(y_test, y_pred_test)}")
```

The model scored better on the training set, with an improvement in 0.083 in adjusted R2. This suggests that there is some degree of overfitting. Let's try some different training sets and see if we get similar results.

```python
# Generate new train/test sets for comparision
y_new = df_grouped['recipe_number'].to_numpy()
x_new = df_grouped.drop(columns=['recipe_number']).to_numpy()

x_train_new, x_test_new, y_train_new, y_test_new = train_test_split(x_new, y_new, test_size=0.2, random_state=0)

best_regressor_new = BaggingRegressor(estimator=DecisionTreeRegressor(max_depth=3, min_samples_split=2), random_state = 0, n_estimators=30)
best_regressor_new.fit(x_train_new, y_train_new)

y_pred_train_new = best_regressor_new.predict(x_train_new)
y_pred_test_new = best_regressor_new.predict(x_test_new)

print(f"Adjusted R2 for 1st New Training Dataset: {r2_score(y_train_new, y_pred_train_new)}")
print(f"Adjusted R2 for 1st New Test Dataset: {r2_score(y_test_new, y_pred_test_new)}")
```

```python
# Generate another pair of train/test sets for comparision
y_new_2 = df_grouped['recipe_number'].to_numpy()
x_new_2 = df_grouped.drop(columns=['recipe_number']).to_numpy()

x_train_new_2, x_test_new_2, y_train_new_2, y_test_new_2 = train_test_split(x_new_2, y_new_2, test_size=0.2, random_state=20)

best_regressor_new_2 = BaggingRegressor(estimator=DecisionTreeRegressor(max_depth=3, min_samples_split=2), random_state = 0, n_estimators=30)
best_regressor_new_2.fit(x_train_new_2, y_train_new_2)

y_pred_train_new_2 = best_regressor_new_2.predict(x_train_new_2)
y_pred_test_new_2 = best_regressor_new_2.predict(x_test_new_2)

print(f"Adjusted R2 for 2nd New Training Dataset: {r2_score(y_train_new_2, y_pred_train_new_2)}")
print(f"Adjusted R2 for 2nd New Test Dataset: {r2_score(y_test_new_2, y_pred_test_new_2)}")
```

There were vastly different results displayed by these 2 additional train/test sets. 

In the first, the model performed much worse. The predictions on the test set went down from 0.805 to 0.737 and the gap between train and test sets went up from 0.083 to 0.131 when compared with the original model.

In the second, the model performed much better. The predictions on the test set went up from 0.805 to 0.897 and the test set actually scored better than the train set. 

Notably, in all 3 models the score on their respective training sets were similar (0.887, 0.868, 0.856). So Bagging is training to pretty consistent levels regardless of the training set. This shows that the major difference between each model is how well they score on the test set. In other words, there is high variance in the model.

Bagging is typically considered as effective in reducing variance/overfitting. Let's theorize why variance is high despite using Bagging.

The most likely reason is that the dataset is too small. Recall that our data cleaning effort kept 16230 samples of the original dataset, but that grouping by recipe was necessary for this experiment. Since the original dataset was taken from 1 website's ranking list from only a single year, this means that the grouped dataset has only 100 samples (1 for each recipe). An 8:2 split was made to this to generate a training set of 80 samples. This training set was further split during 5-fold cross validation, resulting in only 64 samples being used to train the model.

To improve the model, it would be good to increase the dataset in two ways. Firstly, by scraping data for the top 100 lists on www.tasteofhome.com for every year that they have published. Secondly, the scope could be expanded to include other websites' top recipes lists, as focusing on a single website may not be representative of the entire home-chef audience. Not only would both of these methods increase the size of the dataset, but also provide more samples for each point. Our model only trained on 1 recipe of each rank (one #1, one #2, etc...). Having more samples of each rank will help to reduce the strength of outliers.
